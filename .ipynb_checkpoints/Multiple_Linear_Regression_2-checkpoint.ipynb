```python
# import section ------------------------------------------------------------------------------------------------------------- #

# saving the model ----------------------------------------------------------------------------------------------------------- #
import pickle

# relational database and mathematics related libraries ---------------------------------------------------------------------- #
import numpy as np
import pandas as pd
from pandasql import sqldf
import statsmodels.api as sm
from scipy import stats, special

# specific functions and models ---------------------------------------------------------------------------------------------- #
from statsmodels.stats.outliers_influence import variance_inflation_factor

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Testing Model assumptions -------------------------------------------------------------------------------------------------- #
# Tests for heteroscedasticity ----------------------------------------------------------------------------------------------- #
from statsmodels.stats.diagnostic import het_breuschpagan, het_white, het_goldfeldquandt
# Tests for normality -------------------------------------------------------------------------------------------------------- #
from statsmodels.stats.diagnostic import normal_ad, kstest_normal
from statsmodels.stats.stattools import jarque_bera, omni_normtest
# Tests for constant error variance ------------------------------------------------------------------------------------------ #
from statsmodels.stats.diagnostic import acorr_breusch_godfrey
# Tests for autocorrelation -------------------------------------------------------------------------------------------------- #
from statsmodels.stats.stattools import durbin_watson


# libraries for visualization ------------------------------------------------------------------------------------------------ #
import seaborn as sns
import matplotlib.pyplot as plt

# load the data and save it into a dataframe --------------------------------------------------------------------------------- #
all_data_df = pd.read_csv(".\datasets-master-datasciencedojo\Online News Popularity\OnlineNewsPopularity.csv")
all_data_df.drop(columns=["url", " timedelta"], inplace=True)

# intial print to get an overview over the data and it's usability ----------------------------------------------------------- #
#print(all_data_df.head())
print(all_data_df.info())
```

```python
# reduction of correlation matrix to readable size for heatmap --------------------------------------------------------------- #

# threshold of correlation, heatmap does not display values below absolute correlation limit value specified ----------------- #
correlation_threshold = 0.2

corr = all_data_df.corr()

corr_dict = corr.to_dict()

corr_red_dict = dict()

for key in corr_dict.keys():
    t_dct = { k:v for k, v in corr_dict[key].items() if (v != 1.0) & ( (v < -0.2) | (v > 0.2) )}
    corr_red_dict.update({key:t_dct})

corr_red_df = pd.DataFrame.from_dict(corr_red_dict)

corr_red_df.sort_index(axis=0, inplace=True)
corr_red_df.sort_index(axis=1, inplace=True)

corr_red_df

# correlation matrix --------------------------------------------------------------------------------------------------------- #
plt.clf()
plt.figure(figsize=(36,36))
sns.heatmap(corr_red_df, vmin=-1, vmax=1, xticklabels=corr_red_df.columns, yticklabels=corr_red_df.index, cmap="vlag", annot=True, fmt=".2f")
plt.yticks(rotation=30)
plt.show()
```
